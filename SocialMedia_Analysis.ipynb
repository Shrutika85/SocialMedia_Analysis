```python
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
ax=sns.countplot(x='platform',data=df)

for count in ax.containers:
  ax.bar_label(count)
```

```python
(df['gender']=="male").count()
```

```python
plat=df.groupby(['platform']).sum()
plat.head()

ax=sns.barplot(x="platform",y="time_spent",data=plat)
for label in ax.containers:
  ax.bar_label(label)
```

```python
plat=df.groupby(['gender']).sum()
plat.head()

ax=sns.barplot(x="gender",y="time_spent",data=plat)
for label in ax.containers:
  ax.bar_label(label)
```

```python
ax=sns.countplot(x='platform',hue="gender",data=df)
```

```python
ax=sns.countplot(x='platform',hue="interests",data=df)
```

```python
agedf=df.groupby(['age','gender','platform']).sum()
agedf.head()
```

```python
ax=sns.barplot(x='age',hue="platform",data=agedf,y='time_spent')
sns.set(rc={'figure.figsize':(18,5)})
ax.set(xlabel="Age",ylabel="Time_spent")
ax.set_xlim(15, 20)
```
